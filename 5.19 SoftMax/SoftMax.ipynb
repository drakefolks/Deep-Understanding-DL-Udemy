```python
# importing needed libraries
import numpy as np
import torch
import torch.nn as nn # neural networks
import matplotlib.pyplot as plt
```

# Computing softmax manually in numpy

```python
# list of numbers 
z = [4,5,6]

# computing the result of softmax
numerator = np.exp(z)
denominator = np.sum(np.exp(z))
sigma = numerator / denominator

print(sigma,"\n",np.sum(sigma))
```

```python
# calculating softmax with some random numbers
z = np.random.randint(low=-7, high=17, size=27)
print("Z:",z)

# computing the result of softmax
numerator = np.exp(z)
denominator = np.sum(np.exp(z))
sigma = numerator / denominator

print('\nSigma:',sigma)
# comparing 
plt.plot(z,sigma,'ko')
plt.xlabel('Original number (z)')
plt.ylabel('Softmaxed $\sigma$')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

# Softmax in pytorch

```python
# create instance of softmax activation class
softMax = nn.Softmax(dim=0)

# apply the data to the softMax function
sigmaTensor = softMax(torch.Tensor(z)) # must convert z to tensor

# print result tensor
print("Sigma",sigmaTensor)

# plotting results of numpy VS pytorch (should be the same)
plt.plot(sigma,sigmaTensor,'ko')
plt.xlabel('"Manual" softmax')
plt.ylabel('Pytorch nn.softmax')
plt.title(f'Both methods correlate at r={np.corrcoef(sigma,sigmaTensor)[0,1]}')
plt.show()
```

# Goal of SoftMax
- transforms numerical values that have *no intrinsic value* to a **probability distribution** which maps to specific categories that the network is trying to learn
