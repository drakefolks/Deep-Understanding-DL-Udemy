# The T-test
# $t_k = \frac{\bar{x} - \bar{y}}{s/\sqrt{n}} = \frac{\text{Difference of means}}{\text{Standard deviations}}$
## The goal:
<div style='text-align: center;'>
    <ul style='text-align: left; display: inline-block;'>
        <li>compare two different model accuracies to see if there is a meaningful statistical difference</li>
        <li>if P is less than 0.05, there is a meaningful statistical difference between models</li>
    </ul>
</div>


</div>

```python
# importing needed libraries 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# setting up parameters 
n1 = 30 # num samples in dataset 1
n2 = 40 # num smaples in dataset 2
mu1 = 1 # population mean in dataset 1
mu2 = 2 # population mean in dataset 2

# generating random data
data1 = mu1 + np.random.randn(n1) # will be accuracies of model 1
data2 = mu2 + np.random.randn(n2) # will be accuracies of model 2
```

```python
# plotting samples 
plt.plot(np.zeros(n1), data1, 'ro', markerfacecolor='w',markersize=14)
plt.plot(np.ones(n2), data2, 'bs', markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1], labels=['Group 1', 'Group 2'])
plt.show()
```

```python
# t-test with the stats package
t,p = stats.ttest_ind(data1,data2) #_ind = independent samples

print("T:",t)
print("P:",p) # P value should be below 0.05 to be statistically significant 
```

```python
# plotting t-test results
fig = plt.figure(figsize=(10,4))
plt.rcParams.update({'font.size':12})

# this will add random x coordinates to spread out the data in order to reduce overlapping 
plt.plot(0+np.random.randn(n1)/15, data1, 'ro',markerfacecolor='w',markersize=14)
plt.plot(1+np.random.randn(n2)/15, data2, 'bs', markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group 1','Group 2'])

# setting title to include t-value and p-value
plt.title(f't={t:.2f},p={p:.3f}')
plt.show()
```
